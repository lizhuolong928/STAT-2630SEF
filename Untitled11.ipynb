```python

!pip install vaderSentiment

from pyspark.sql.functions import udf, col, to_date, date_format, when
from pyspark.sql.types import IntegerType, StringType
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
# emotion tagging UDFпјҲreturn 0/1/2пјү
def get_sentiment(text):
    if text is None:
        return 1  # neutral
    score = analyzer.polarity_scores(text)['compound']
    if score <= -0.05:
        return 0  # negative
    elif score >= 0.05:
        return 2  # positive
    else:
        return 1  # neutral
sentiment_udf = udf(get_sentiment, IntegerType())
```

```python
def has_topic(text, keywords):
    if text is None:
        return 0
    text_lower = text.lower()
    for kw in keywords:
        if kw in text_lower:
            return 1
    return 0
```

```python
job_keywords = ['job', 'work', 'employ', 'unemployment', 'post', 'replace', 'supersede']
privacy_keywords = ['privacy', 'data leak', 'personal information', 'leakage', 'intimacy', 'monitoring']
bias_keywords = ['bias', 'discriminat', 'discriminate', 'unfair']
reliability_keywords = ['reliable', 'accuracy', 'mistake', 'reliable', 'accuracy', 'misdiagnose', 'accident']
standard_keywords = ['standard', 'regulation', 'measure', 'norm', 'supervise', 'legislation']
```

```python
job_udf = udf(lambda t: has_topic(t, job_keywords), IntegerType())
privacy_udf = udf(lambda t: has_topic(t, privacy_keywords), IntegerType())
bias_udf = udf(lambda t: has_topic(t, bias_keywords), IntegerType())
reliability_udf = udf(lambda t: has_topic(t, reliability_keywords), IntegerType())
standard_udf = udf(lambda t: has_topic(t, standard_keywords), IntegerType())
```

```python

df = df.withColumn("sentiment", sentiment_udf(col("text")))
df = df.withColumn("topic_job", job_udf(col("text")))
df = df.withColumn("topic_privacy", privacy_udf(col("text")))
df = df.withColumn("topic_bias", bias_udf(col("text")))
df = df.withColumn("topic_reliability", reliability_udf(col("text")))
df = df.withColumn("topic_standard", standard_udf(col("text")))

# Extract the monthпјҲpublish_time the format might be "2026-04-14 12:31:04" or similarпјү
df = df.withColumn("date", to_date(col("publish_time")))
df = df.withColumn("month", date_format(col("date"), "yyyy-MM"))
```

```python
from pyspark.ml.feature import Tokenizer, StopWordsRemover, HashingTF, IDF, StringIndexer
from pyspark.ml import Pipeline

tokenizer = Tokenizer(inputCol="text", outputCol="words")
stopwords = StopWordsRemover.loadDefaultStopWords("english") + ["is","is","is"]
remover = StopWordsRemover(inputCol="words", outputCol="filtered", stopWords=stopwords)
hashingTF = HashingTF(inputCol="filtered", outputCol="rawFeatures", numFeatures=5000)
idf = IDF(inputCol="rawFeatures", outputCol="features")

feature_pipeline = Pipeline(stages=[tokenizer, remover, hashingTF, idf])
feature_model = feature_pipeline.fit(df)
df_featured = feature_model.transform(df)
```

```python
# ==================== 1. install dependencies ====================
!pip install vaderSentiment pyspark -q

# ==================== 2. initialize Spark ====================
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("AIGovernance").getOrCreate()

# ==================== 3. upload JSON data ====================
df1 = spark.read.json("/content/data1.json")
df2 = spark.read.json("/content/data2.json")
df = df1.union(df2)
print(f"жҖ»иЎҢж•°: {df.count()}")

# ==================== 4. automatically label emotions and topics ====================
from pyspark.sql.functions import udf, col, to_date, date_format, when
from pyspark.sql.types import IntegerType
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
def get_sentiment(text):
    if text is None:
        return 1
    score = analyzer.polarity_scores(text)['compound']
    if score <= -0.05:
        return 0
    elif score >= 0.05:
        return 2
    else:
        return 1
sentiment_udf = udf(get_sentiment, IntegerType())
df = df.withColumn("sentiment", sentiment_udf(col("text")))
df = df.withColumn("sentiment_text",
                   when(col("sentiment") == 0, "negative")
                   .when(col("sentiment") == 1, "neutral")
                   .when(col("sentiment") == 2, "positive"))

def has_topic(text, keywords):
    if text is None:
        return 0
    text_lower = text.lower()
    return 1 if any(kw in text_lower for kw in keywords) else 0

job_keywords = ['job', 'work', 'employ', 'unemployment', 'post', 'replace', 'supersede']
privacy_keywords = ['privacy', 'data leak', 'personal information', 'leakage', 'intimacy', 'monitoring']
bias_keywords = ['bias', 'discriminat', 'discriminate', 'unfair']
reliability_keywords = ['reliable', 'accuracy', 'mistake', 'reliable', 'accuracy', 'misdiagnose', 'accident']
standard_keywords = ['standard', 'regulation', 'measure', 'norm', 'supervise', 'legislation']


topic_mappings = [
    ("topic_job", job_kw),
    ("topic_privacy", privacy_kw),
    ("topic_bias", bias_kw),
    ("topic_reliability", reliability_kw),
    ("topic_standard", standard_kw)
]


for col_name, kw_list in topic_mappings:
    udf_func = udf(lambda t, kw=kw_list: has_topic(t, kw), IntegerType())
    df = df.withColumn(col_name, udf_func(col("text")))

df = df.withColumn("date", to_date(col("publish_time")))
df = df.withColumn("month", date_format(col("date"), "yyyy-MM"))

# ==================== 5. feature engineering ====================
from pyspark.ml.feature import Tokenizer, StopWordsRemover, HashingTF, IDF, StringIndexer
from pyspark.ml import Pipeline

tokenizer = Tokenizer(inputCol="text", outputCol="words")
stopwords = StopWordsRemover.loadDefaultStopWords("english") + ["is","is","is","here","with","have","me"]
remover = StopWordsRemover(inputCol="words", outputCol="filtered", stopWords=stopwords)
hashingTF = HashingTF(inputCol="filtered", outputCol="rawFeatures", numFeatures=5000)
idf = IDF(inputCol="rawFeatures", outputCol="features")

feature_pipeline = Pipeline(stages=[tokenizer, remover, hashingTF, idf])
feature_model = feature_pipeline.fit(df)
df_featured = feature_model.transform(df)

# ==================== 6. classification of love songs ====================
indexer = StringIndexer(inputCol="sentiment_text", outputCol="label")
indexer_model = indexer.fit(df_featured)
df_indexed = indexer_model.transform(df_featured)

train, test = df_indexed.randomSplit([0.8, 0.2], seed=42)

from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

lr = LogisticRegression(featuresCol="features", labelCol="label", maxIter=20, regParam=0.1)
lr_model = lr.fit(train)
lr_pred = lr_model.transform(test)

evaluator = MulticlassClassificationEvaluator(labelCol="label", metricName="f1")
f1_score = evaluator.evaluate(lr_pred)
print(f"logistic regression weightinhg F1 = {f1_score:.4f}"

# ==================== 7. detailed assessmentпјҲmanual calculation F1пјҢavoid API questionпјү ====================
from pyspark.mllib.evaluation import MulticlassMetrics
pred_rdd = lr_pred.select("label", "prediction").rdd.map(lambda row: (float(row.prediction), float(row.label)))
metrics = MulticlassMetrics(pred_rdd)
labels = indexer_model.labels

print(f"\n the number of training samples: {train.count()}")
print(f" number of test samples : {test.count()}")
for i, label in enumerate(labels):
    prec = metrics.precision(i)
    rec = metrics.recall(i

    # manual calculation F1
    if prec + rec > 0:
        f1 = 2 * (prec * rec) / (prec + rec)
    else:
        f1 = 0.0
    print(f"{label:10s} -> accurate rate : {prec:.4f}, racall rate : {rec:.4f}, F1: {f1:.4f}")
print(f" weighted average  F1 = {metrics.weightedFMeasure():.4f}"

# ==================== 8. export visualization data ====================
topic_cols = ["topic_job", "topic_privacy", "topic_bias", "topic_reliability", "topic_standard"]
output_df = lr_pred.select(
    col("sentiment_text"),
    col("month"),
    *topic_cols
)
# merge into a single one CSV file
output_df.coalesce(1).write.csv("/content/results/for_visualization.csv", header=True, mode="overwrite")
print("\nвң… the visualized data has been exported to  /content/results/for_visualization.csv")
print("   fieldпјҡsentiment_text, month, topic_job, topic_privacy, topic_bias, topic_reliability, topic_standard"
```
